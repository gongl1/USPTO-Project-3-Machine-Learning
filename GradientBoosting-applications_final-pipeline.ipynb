Data Pre-Processing

```python
# Since there are 15million+ rows, I use a sample of 0.002 of the 15million+ rows 
df_selected_final_sample = df_selected_final.sample(frac=.002)
df_selected_final_sample.info()
```

```python
df_selected_final_sample.loc[(df_selected_final_sample['uspc_class']) == '074', :]
```

```python
X = df_selected_final_sample.iloc[:,1:].astype('str')
X
```

```python
y = df_selected_final_sample["appl_status_desc"]
y
```

```python
# impute-replace missing data
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

from sklearn.datasets import make_hastie_10_2
from sklearn.ensemble import GradientBoostingClassifier

from sklearn.feature_selection import SelectFromModel

from sklearn.model_selection import train_test_split

# In this case, we only take care of categorical_columns since it only has categorical data
categorical_columns = ['application_invention_type', 'uspc_class', 'uspc_subclass', 'invention_title', 'small_entity_indicator']
numerical_columns = []

clf = GradientBoostingClassifier(n_estimators=500, learning_rate=0.1, max_depth=3, random_state=0)

# Split X, y to test and train
X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=1)
```

```python
categorical_pipe = Pipeline([
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])
numerical_pipe = Pipeline([
    ('imputer', SimpleImputer(strategy='median'))
])
preprocessing = ColumnTransformer(
    [('cat', categorical_pipe, categorical_columns),
#      ('num', numerical_pipe, numerical_columns)
    ],
)

# above are transformers to data
rf = Pipeline([
    ('preprocess', preprocessing),
    ('feature_selection', SelectFromModel(clf, prefit=False)),
    ('classifier', clf)
])
```

```python
rf.fit(X_train,y_train)
rf.score(X_test,y_test)
# rf is our ML model here
```

```python
# Score the model
print(f"Training Data Score: {rf.score(X_train, y_train)}")
print(f"Testing Data Score: {rf.score(X_test, y_test)}")
```
