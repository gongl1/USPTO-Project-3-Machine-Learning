Data Pre-Processing

```python
# Since there are 15million+ rows, I use a sample of 0.002 of the 15million+ rows 
df_selected_final_sample = df_selected_final.sample(frac=.002)
df_selected_final_sample.info()
```

```python
# Convert categorical data into a numerical form and our machine learning algorithm can take in that as input.
test = pd.get_dummies(df_selected_final_sample, columns=['application_invention_type', 'uspc_class', 'uspc_subclass', 'invention_title', 'small_entity_indicator'])
```

```python
X = test
y = df_selected_final_sample["appl_status_desc"]
print(X.shape, y.shape)
```

```python
X
```

```python
X = X.iloc[:,1:]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from tensorflow.keras.utils import to_categorical
```

```python
# Split X, y to test and train
X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=1)
```

```python
# Get a general score of the model
from sklearn.ensemble import ExtraTreesClassifier

model = ExtraTreesClassifier(n_jobs=-1)
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

# Train the Model
Using a logistic regression


```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=3000, criterion='entropy', max_depth=5)

# Train the model
rf = rf.fit(X_train, y_train)
```

```python
# Score the model
print(f"Training Data Score: {rf.score(X_train, y_train)}")
print(f"Testing Data Score: {rf.score(X_test, y_test)}")
```

# Predict

```python
predictions = rf.predict(X_test)
df_pred = pd.DataFrame({"Actual":y_test, "Predicted":predictions}) 
# pd.set_option("display.max_rows", None, "display.max_columns", None)
df_pred
```
